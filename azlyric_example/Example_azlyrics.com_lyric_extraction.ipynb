## Example of Scrapping lyric of one song from a certain artist from azlyrics.com

```python
import requests
import re
```

## Here I am choosing the artist as Eminem

```python
url = 'https://www.azlyrics.com/e/eminem.html'
http_request = requests.get(url)  # gets the html page of emimem from azlyrics.com
print(http_request.text)
```

## Extracting the links of all songs of artist from the html page

```python
links = re.findall('href="..(/lyrics.*?")',http_request.text) 
```

```python
links 
```

## Adding in the prefix https://www.azlyrics.com and removing the last character (") from each link so that the link can be used for extracting lyric

```python
links_cleaned = []
for link in links:
    links_cleaned.append('https://www.azlyrics.com' + link[:-1])
```

```python
links_cleaned
```

## Extracting the html page of the first song (https://www.azlyrics.com/lyrics/eminem/infinite.html) to extract lyrics
#### This is just an example to extract the lyric for 1 song, same logic can be applied in a for loop to get all song lyrics

```python
eminem_first_song = requests.get(links_cleaned[0]).text
print(eminem_first_song)
```

```python
from bs4 import BeautifulSoup
```

```python
eminiem_soup = BeautifulSoup(eminem_first_song, 'html.parser')
```

```python
song = eminiem_soup.body.find(name='div',class_ = "col-lg-8").text.splitlines()
```

```python
song
```

```python
song = list(filter(lambda a: a != '', song)) # removes the unwanted '' in the list
song
```

## The previous steps gets all the text in the webpage but we need to extract only the lyric part
### In azlyrics there is a Submit Corrections button before which you have the end of the lyric and the first 3 lines are of the title and artist name which we don't want as part of lyrics 

```python
temp = song.index(' Submit Corrections')  # Extracting only lyric part
song[3:temp]
```
